```python
import xarray as xr
ds = xr.open_dataset('ocean_ds.nc')
print(ds)
```

```python
ds.head(25)
```

```python
import pandas as pd
df = ds.to_dataframe().reset_index()
print(df.head())
```

```python
import numpy as np

def detect_outz(data , threshold=3) :
    z_scores = (data - data.mean()) / data.std()
    return np.abs(z_scores) > threshold

outliers_time = detect_outz(df['time'])
print(f"total outliers:{outliers_time.count()}")
```

```python
#the erroe showing above about the time is handled here
ds['time']
ds= ds.dropna(dim='time')

```

```python
#now we will handle missing values that is interpolate
df_clean = df[~outliers_time]
df_clean = df_clean.reset_index(drop=True)
df_inp = df_clean.interpolate(method='linear', limit_direction='forward' , axis=0)
```

```python
#normalizing
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
column_to_normalize = ['time']

df_inp[column_to_normalize] = scaler.fit_transform(df_inp[column_to_normalize])
```

```python
#visualizing the data cleaned one
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df_inp[column_to_normalize])
plt.suptitle("cleaned and normalized data " , y=1.02)
plt.show()
```

```python
import numpy as np

def detect_outliers_zscore(data, threshold=3):
    z_scores = (data - data.mean()) / data.std()
    return np.abs(z_scores) > threshold


outliers_temp = detect_outliers_zscore(df['longitude'])
outliers_salinity = detect_outliers_zscore(df['latitude'])

# Combine outliers
outliers_combined = outliers_temp | outliers_salinity

print(f"Total outliers: {outliers_combined.sum()}")
```

```python
df_cleaned = df[~outliers_combined]
df_cleaned = df_cleaned.reset_index(drop=True)

```

```python
df_interpolated = df_cleaned.interpolate(method='linear', limit_direction='forward', axis=0)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
columns_to_normalize = ['longitude', 'latitude']  # Choose numeric columns

df_interpolated[columns_to_normalize] = scaler.fit_transform(df_interpolated[columns_to_normalize])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(df_interpolated[columns_to_normalize])
plt.suptitle("Cleaned and Normalized Marine Sensor Data", y=1.02)
plt.show()
```

```python
df_interpolated.to_csv("preprocessed_marine_data.csv", index=False)
print("Data saved to 'preprocessed_marine_data.csv'")
```
