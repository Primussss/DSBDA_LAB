```
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
df
```

```
import seaborn as sns
sns.displot(x=df['petal_width'], bins=10)
sns.displot(df['petal_length'], bins=10 , kde=False)
```

```
sns.jointplot (x=df['petal_length'], y=df['sepal_length'], kind='scatter')
sns.jointplot (x=df['petal_length'], y=df['sepal_length'], kind='hex')
```

```
sns.rugplot(df['sepal_length'])
```

```
#now we have categorical plots : 1.bar plot 2.count plot 3.box plot 4. voilin plot
import numpy as np

sns.barplot(x=df['species'], y=df['sepal_length'])

```

```
sns.barplot(x=df['species'], y=df['sepal_length'], estimator=np.std)
```

```
sns.countplot(x=df['species'], data=df)
```

```
sns.boxplot(x=df['species'], y=df['sepal_length'])
```

```
sns.boxplot(x=df['species'], y=df['sepal_length'], hue=df['species'])
```

```
sns.violinplot(x=df['species'], y=df['sepal_length'] , data=df)
```

```
sns.violinplot(x=df['species'], y=df['sepal_length'] , data=df, hue=df['species'])
```

```
#now we will plot advance plots and matrix plot ad(a. Strip Plot ,b. Swarm Plot) ma(a. Heat Map ,b. Cluster Map)
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('iris')
sns.stripplot(x=df['species'], y=df['sepal_length'] , data=df , jitter=False)
```

```
sns.stripplot(x=df['species'], y=df['sepal_length'] , data=df , jitter=True)
```

```
sns.stripplot(x=df['species'], y=df['sepal_length'] , data=df , jitter=True, hue=df['species'])
```

```
sns.swarmplot(x=df['species'], y=df['sepal_length'] , data=df)
```
